# Scipy stats

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

## Scipy stats

the functions are as follows:

- `pdf` Probability Density Function
- `pmf` Probability mass function for discrete rv
- `ppf` Percent point function. Basically returns the inverse of the cdf
- `cdf` Cumulative distribution function
- `sf` Survival Function $SF_X(t) = P(X>t)$ or $SF_X(t) = 1 - F_X(t)$
- `logpmf` This gives you the surprisal of the event at different points where PMF is defined.


```python
fig, axes = plt.subplots(1,2,figsize=(16, 9 ),dpi = 300)
dist = stats.bernoulli

p = 0.8

x = np.linspace(0,1,10,endpoint=True)
rv = dist.pmf(x,p)
rvc = dist.cdf(x,p)

axes[0].plot(x,rv,'o', ms =4)
axes[1].plot(x,rvc)
```

```python
fig, axes = plt.subplots(1,2,figsize=(11,9),dpi = 300)
dist2 = stats.binom

n = 100
p = 0.5

# Binomial Distribution
x2 = np.arange(dist2.ppf(0, n, p), dist2.ppf(1, n, p))
rv2 = dist2.pmf(x2, n , p)
rv2c = dist2.cdf(x2, n, p)

axes[0].plot(x2,rv2,'ko', ms = 3)
axes[1].plot(x2,rv2c)
```

```python
fig, axes = plt.subplots(1,2,figsize=(16,9),dpi = 300)
dist = stats.poisson

mu = 9

# Poisson Distribution
x = np.arange(dist.ppf(0, mu), dist.ppf(0.9999999, mu))
rv = dist.pmf(x, mu)
rvc = dist.cdf(x, mu)

axes[0].plot(x,rv,'k', ms = 3)
axes[1].plot(x,rvc)
```
